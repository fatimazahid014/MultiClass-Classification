###  Tuning max_features

```python
#@title Importing modules
import pandas as pd
import numpy as np
```

```python
#@title Loading data

glass_data = pd.read_csv\
('https://raw.githubusercontent.com/fenago/DSBook/main/Chapter%204/glass.csv')

glass_data.head()
```

```python
#@title Remove the Type column using .drop() and then extract the Type targetvariable
target = glass_data.pop('Type')
glass_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
#@title Train_test_split

X_train, X_test, y_train, y_test = train_test_split\
                                 (glass_data, target, test_size=0.4, \
                                 random_state=188)
```

```python
#@title Instantiate RandomForestClassifier with random_state=42 and n_estimators=30, max_depth=2, min_samples_leaf=7, max_features=4, and then fit the model with the training set:
from sklearn.ensemble import RandomForestClassifier
# ---- type your code here ---#



# --- Solution --- #
f_model = RandomForestClassifier(random_state=42, \
                                  n_estimators=30, max_depth=2,min_samples_leaf=7,max_features=8)
f_model.fit(X_train, y_train)
```

```python
#@title Make predictions on the training and testing sets with .predict()

train_preds = f_model.predict(X_train)
test_preds = f_model.predict(X_test)
```

```python
#@title Accuracy Scores
from sklearn.metrics import accuracy_score

train_acc = accuracy_score(y_train, train_preds)
test_acc = accuracy_score(y_test, test_preds)
```

```python
#@title Print the accuracy scores: train_acc and test_acc:
print(train_acc)
print(test_acc)
```

```python
#@title Instantiating another RandomForestClassifier with random_state=42, n_estimators=30, and max_depth=2, max_features=0.2, and min_samples_leaf=7
f_model2 = RandomForestClassifier(random_state=42, \
                                   n_estimators=30,\
                                    max_depth=2,\
                                       min_samples_leaf=7,\
                                          max_features=0.9)
f_model2.fit(X_train, y_train)
```

```python
#@title Make predictions on the training and testing sets with .predict()
train_preds2 = f_model2.predict(X_train)
test_preds2 = f_model2.predict(X_test)
```

```python
#@title Calculating the accuracy score for the training and testing sets
train_acc2 = accuracy_score(y_train, train_preds2)
test_acc2 = accuracy_score(y_test, test_preds2)
```

```python
#@title Printing the accuracy scores: train_acc and test_acc:
print(train_acc2)
print(test_acc2)
```
