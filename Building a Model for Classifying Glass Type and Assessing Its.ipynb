### Building a Model for Classifying Glass Type and Assessing Its Performance

### Import the pandas package:

```python
#@title Import libraries
import pandas as pd
import numpy as np
```

```python
#@title Loading the "glass" data:
glass_data = pd.read_csv\
('https://raw.githubusercontent.com/fenago/DSBook/main/Chapter%204/glass.csv')

glass_data.head()
```

```python
#@title Extracting the 'Type' column using the .pop() method from pandas to predict the type of glass

target = glass_data.pop('Type')
glass_data.head()
```

```python
#@title Importing the train_test_split function from sklearn.model_selection
from sklearn.model_selection import train_test_split
```

```python
#@title Splitting the dataset into training and testing sets with the df, y, test_size=0.4, and random_state=188 parameters

X_train, X_test, y_train, y_test = train_test_split\
                                  (glass_data, target, test_size=0.4, \
                                  random_state=188)
```

```python
#@title Import RandomForestClassifier from sklearn.ensemble:
from sklearn.ensemble import RandomForestClassifier

f_model = RandomForestClassifier(random_state=80, \
                                  n_estimators=10)
```

```python
#@title Fitting RandomForestClassifier with the training set
f_model.fit(X_train, y_train)
```

```python
#@title Predicting the outcome of the training set with the .predict()method

train_preds = f_model.predict(X_train)
train_preds
```

```python
#@title Importing the accuracy_score function from sklearn.metrics:
from sklearn.metrics import accuracy_score
```

```python
#@title Default title text
train_acc = accuracy_score(y_train, train_preds)
print(train_acc)
```

```python
#@title Predicting the outcome of the testing set with the .predict() method
test_preds = f_model.predict(X_test)
```

```python
#@title Accuracy:
test_acc = accuracy_score(y_test, test_preds)
print(test_acc)
```
